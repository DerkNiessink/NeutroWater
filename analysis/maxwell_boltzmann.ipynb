```python
import matplotlib.pyplot as plt
import scienceplots
plt.style.use("science")
import numpy as np
import seaborn as sns
import scipy.integrate as spi


from neutrons.models.maxwell_boltzmann import MaxwellBoltzmann
```

```python
mw = MaxwellBoltzmann()
velocities = [v for v in np.arange(0, 10000, 0.1)]
densities = [mw.distribution(v) for v in velocities]
plt.figure(figsize=(5, 4))

J = 6.24150907e18 
energies = [0.5 * mw.m * v ** 2 * J for v in velocities]
plt.plot(energies, densities, linewidth=1.4, label = "Boltzmann distribution")
plt.xlabel("$E (eV)$", fontsize=14)
plt.ylabel(r"$P(E)$", fontsize=14)

plt.xlim(-0.01,0.3)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.ylim(0, 0.0005)
plt.axvline(0.25, color='black', lw=0.8, linestyle='--', label = "Threshold energy (0.252 eV)")
plt.ylim(0)
plt.legend(fontsize=13, loc = "upper left")
plt.show()
```

```python
sampled_points= [mw.thermal_velocity() for _ in range(10000)]

plt.figure(figsize=(7, 5))
plt.plot(velocities,densities, "-", linewidth=1.5, color = "k", label = "Maxwell-Boltzmann Distribution")
sns.histplot(sampled_points, bins=50, stat="density", label = "Sampled Velocities")
plt.xlabel("$v (m/s)$", fontsize=15)
plt.ylabel(r"Density", fontsize=15)
plt.legend()
plt.title("Sampled Thermal Neutron Energies", fontsize=15)
plt.show()
```

```python
sampled_points= [mw.thermal_energy() for _ in range(100000)]

plt.figure(figsize=(7, 5))
sns.histplot(sampled_points, bins=100, stat="density")
plt.xlabel("Neutron Energy (eV)", fontsize=15)
plt.ylabel(r"Density", fontsize=15)
plt.title("Sampled Thermal Neutron Energies", fontsize=15)
plt.xlim(-0.005, 0.25)
plt.show()
```
